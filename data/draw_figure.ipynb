```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
plt.rcParams['figure.dpi'] = 300
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 5}
plt.rc('font', **font)
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['lines.linewidth']=0.7
plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'
plt.rcParams['xtick.major.width']=0.4
plt.rcParams['ytick.major.width']=0.4
plt.rcParams['xtick.major.size']=2
plt.rcParams['ytick.major.size']=2
```

```python
fig,ax = plt.subplots(1,3,figsize=(7.0,7.0*4/16))
# ax.spines['right'].set_visible(False)
# ax.spines['top'].set_visible(False)

w = pd.read_csv('15.csv',names=['time','left','right'])[1000:1500]
time,left,right = np.array(w['time']),np.array(w['left']),np.array(w['right'])
right_ref = np.clip(0.15*np.sin(2*np.pi/0.84*(time-2.1))-0.03,0,200) 
left_ref = np.clip(0.15*np.sin(np.pi+2*np.pi/0.84*(time-2.1))-0.03,0,200) 

ax[0].plot(time-2.1,left_ref,'--',c='cyan')
ax[0].plot(time-2.1,right_ref,'--',c='orange')
ax[0].plot(time-2.1,left,'b')
ax[0].plot(time-2.1,right,'r')

leg = ax[0].legend(["Left Ref","Right Ref","Left",'Right'],loc=1,ncol=2)
leg.get_frame().set_linewidth(0.1)
ax[0].set_xlim([0,0.84*2.5])
ax[0].set_ylim([0,0.28])
ax[0].set_title('(a) $h-\Delta h=0.12, \Delta h/h=0.2$',y=0.95,loc='left',fontsize=7)
ax[0].set_xlabel('Time (s)',fontsize=7)
ax[0].set_ylabel('Foot Height (m)',fontsize=7)
ax[0].xaxis.set_label_coords(.5,-0.15)
ax[0].yaxis.set_label_coords(-.15,.5)

# better add the reference

# w = pd.read_csv('20.csv',names=['time','left','right'])[2800:3200]
# time,left,right = np.array(w['time']),np.array(w['left']),np.array(w['right'])
# right_ref = np.clip(0.20*np.sin(2*np.pi/0.84*(time-9.25))-0.04,0,200) 
# left_ref = np.clip(0.20*np.sin(np.pi+2*np.pi/0.84*(time-9.25))-0.04,0,200) 

# ax[0][1].plot(time-9.2,left_ref,'--',c='cyan')
# ax[0][1].plot(time-9.2,right_ref,'--',c='orange')
# ax[0][1].plot(time-9.2,left,'b')
# ax[0][1].plot(time-9.2,right,'r')

# leg = ax[0][1].legend(["Left Ref",'Right Ref',"Left",'Right'],loc=1,ncol=2)
# leg.get_frame().set_linewidth(0.1)
# ax[0][1].set_xlim([0,0.84*2.5])
# ax[0][1].set_ylim([0,0.28])
# ax[0][1].set_title('(b) $h-\Delta h=0.16, \Delta h/h=0.2$',y=0.95,loc='left',fontsize=4)
# ax[0][1].set_xlabel('Time (s)',fontsize=4)
# ax[0][1].set_ylabel('Foot Height (m)',fontsize=4)
# ax[0][1].xaxis.set_label_coords(.5,-0.15)
# ax[0][1].yaxis.set_label_coords(-.15,.5)

w = pd.read_csv('25.csv',names=['time','left','right'])[1200:1400]
time,left,right = np.array(w['time']),np.array(w['left']),np.array(w['right'])
right_ref = np.clip(0.25*np.sin(2*np.pi/0.84*(time-5.05))-0.05,0,200) 
left_ref = np.clip(0.25*np.sin(np.pi+2*np.pi/0.84*(time-5.05))-0.05,0,200) 

ax[1].plot(time-5.05,left_ref,'--',c='cyan')
ax[1].plot(time-5.05,right_ref,'--',c='orange')
ax[1].plot(time-5.05,left,'b')
ax[1].plot(time-5.05,right,'r')
leg = ax[1].legend(["Left Ref", "Right Ref", "Left",'Right'],loc=1,ncol=2)
leg.get_frame().set_linewidth(0.1)
ax[1].set_xlim([0,0.84*2.5])
ax[1].set_ylim([0,0.28])
ax[1].set_title('(b) $h-\Delta h=0.20, \Delta h/h=0.2$',y=0.95,loc='left',fontsize=7)
ax[1].set_xlabel('Time (s)',fontsize=7)
ax[1].set_ylabel('Foot Height (m)',fontsize=7)
ax[1].xaxis.set_label_coords(.5,-0.15)
ax[1].yaxis.set_label_coords(-.15,.5)

w = pd.read_csv('15.csv',names=['time','left','right'])[1100:1500]
time,left,right = np.array(w['time']),np.array(w['left']),np.array(w['right'])
vz_r = np.diff(right)/0.03
plt.plot(right[:-1],vz_r,linewidth=0.4)

# w = pd.read_csv('20.csv',names=['time','left','right'])[2800:3200]
# time,left,right = np.array(w['time']),np.array(w['left']),np.array(w['right'])
# vz_r = np.diff(right)/0.03
# plt.plot(right[:-1],vz_r,linewidth=0.4)

w = pd.read_csv('25.csv',names=['time','left','right'])[1200:1600]
time,left,right = np.array(w['time']),np.array(w['left']),np.array(w['right'])
vz_r = np.diff(right)/0.03
plt.plot(right[:-1],vz_r,linewidth=0.4)
plt.ylim([-3.4,4.5])

plt.xlabel('Right Foot Height (m)',fontsize=7)
plt.ylabel('Right Foot $v_z$ (m/s)',fontsize=7)
ax[2].xaxis.set_label_coords(.5,-0.15)
ax[2].yaxis.set_label_coords(-.1,.5)
plt.title('(c) $v_z$ versus $h$',y=0.95,loc='left',fontsize=7)
leg = plt.legend(['$h-\Delta h=0.12$',
#             '$h-\Delta h=0.16$',
            '$h-\Delta h=0.20$'],loc=1)
leg.get_frame().set_linewidth(0.1)

plt.subplots_adjust(wspace=0.26)

plt.savefig(r'footheight.pdf',dpi=300,bbox_inches='tight',pad_inches=0)
```
