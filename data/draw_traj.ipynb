```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
plt.rcParams['figure.dpi'] = 300
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 5}
plt.rc('font', **font)
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['lines.linewidth']=0.7
plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'
plt.rcParams['xtick.major.width']=0.4
plt.rcParams['ytick.major.width']=0.4
plt.rcParams['xtick.major.size']=2
plt.rcParams['ytick.major.size']=2
```

```python
w = pd.read_csv('v033.csv',names=['time','x','y'])[606:]
time,x,y = np.array(w['time']),np.array(w['x']),np.array(w['y'])
plt.figure(figsize=(12,3))
plt.plot(time)
```

```python
fig,ax = plt.subplots(1,1,figsize=(3.5,3.5*9/16))
# ax.spines['right'].set_visible(False)
# ax.spines['top'].set_visible(False)

w06 = pd.read_csv('v06.csv',names=['time','x','y'])[1271:]
w03 = pd.read_csv('v03.csv',names=['time','x','y'])
w42 = pd.read_csv('v42.csv',names=['time','x','y'])[1676:]
time,x06,y06 = np.array(w06['time']),np.array(w06['x']),np.array(w06['y'])
time,x03,y03 = np.array(w03['time']),np.array(w03['x']),np.array(w03['y'])
time,x42,y42 = np.array(w42['time']),np.array(w42['x']),np.array(w42['y'])

ax.plot(x06[::60],y06[::60],linestyle='',c='k',marker='^',ms=3)
ax.plot(x03[::60],y03[::60],linestyle='',c='k',marker='o',ms=3)
ax.plot(x42[::60],y42[::60],linestyle='',c='k',marker='s',ms=3)

ax.plot(x06,y06,'--k')
ax.plot(x03,y03,'--k')
ax.plot(x42,y42,'--k')

leg = ax.legend(["$v_x=0.6,v_y=0.0$","$v_x=0.0,v_y=0.3$","$v_x=-0.4, v_y=-0.2$"],loc=1)
leg.get_frame().set_linewidth(0.1)
# plt.axis('equal')
# ax.set_xlim([-8,12])
# ax.set_ylim([0,0.28])
# ax.set_title('Motion Trajectory',y=0.95,loc='left',fontsize=7)
ax.set_xlabel('$x$ (m)',fontsize=7)
ax.set_ylabel('$y$ (m)',fontsize=7)
# ax.xaxis.set_label_coords(.5,-0.15)

plt.savefig(r'traj.pdf',dpi=300,bbox_inches='tight',pad_inches=0)
```
